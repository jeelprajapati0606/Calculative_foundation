# Part A: Vector & Matrix Fundamentals



###  Que 1 : Represent each student’s subject scores as a vector.

```python
# student 1 score vector
student_1 = data.loc[0, ['math_score', 'science_score', 'english_score']].to_numpy(dtype=float)


# student 2 score vector
student_2 = data.loc[1, ['math_score', 'science_score', 'english_score']].to_numpy(dtype=float)


# student 3 score vector
student_3 = data.loc[2, ['math_score', 'science_score', 'english_score']].to_numpy(dtype=float)


print("Student 1 : ",student_1)
print("Student 2 : ",student_2)
print("Student 3 : ",student_3)
```

## Interpretation :

- Student 1 has the lowest scores, showing weaker performance in all three subjects.

- Student 2 performs moderately well, clearly stronger than Student 1 but not at the top.

- Student 3 excels with high scores across subjects, emerging as the best performer overall.

### Que 2 : Compute L1 Norm and L2 Norm

```python
# L1 Norm
l1_student1 = np.linalg.norm(student_1, ord=1)
l1_student2 = np.linalg.norm(student_2, ord=1)
l1_student3 = np.linalg.norm(student_3, ord=1)


# L2 Norm
l2_student1 = np.linalg.norm(student_1)
l2_student2 = np.linalg.norm(student_2)
l2_student3 = np.linalg.norm(student_3)

print("Student:1 L1 norm: ",l1_student1)
print("Student:2 L1 norm: ",l1_student2)
print("Student:3 L1 norm: ",l1_student3)

print()

print("Student:1 L2 norm: ",l2_student1)
print("Student:2 L2 norm: ",l2_student2)
print("Student:3 L2 norm: ",l2_student3)

```

## Interpretation :

- L1 Norm shows the total magnitude of scores; Student 3 has the highest (259.0), meaning overall stronger cumulative performance.

- L2 Norm reflects the Euclidean distance (overall strength); again, Student 3 leads (≈149.96), confirming their dominance in all subjects.

- Comparison indicates that Student 1 is weakest, Student 2 is moderate, and Student 3 consistently outperforms both in cumulative and balanced score strength.

## Que 3 : Compute Dot Product & Find Angle Between Two Vectors

```python
dot_product = np.dot(student_1, student_2)

print("Dot Product = ",dot_product )

cos_theta = np.dot(student_1, student_2) / (
    np.linalg.norm(student_1) * np.linalg.norm(student_2)
)

angle = np.degrees(np.arccos(cos_theta))

print("\nAngle =",angle)
```

## Interpretation :

- The dot product value (9964.64) measures similarity between Student 1 and Student 2’s score vectors — a higher value indicates stronger alignment in their performance patterns.

- This positive and relatively large result suggests that Student 1 and Student 2 have moderately correlated strengths across math, science, and English, though Student 2 scores higher overall.

- The angle between Student 1 and Student 2’s score vectors is about 3.81°, which is very small.



## Que 5 : Compute Cross Product

```python
cross_product = np.cross(student_1, student_2)

print("Cross Product = ",cross_product )
```

## Interpretation :

- The cross product result [-331.68, 516.24, -253.28] represents a vector perpendicular to both Student 1 and Student 2’s score vectors.
- The non-zero result highlights variation in subject strengths, revealing subtle mismatches in how each student balances math, science, and English.

## Que 6 : Find the projection of one vector onto another.


```python
projection = (np.dot(student_1, student_2) / np.dot(student_2, student_2)) * student_2

print("Projection of Student_1 on Student_2\n", projection)
```

## Interpretation :

- The projection vector [46.85, 55.96, 52.71] shows how Student 1’s performance aligns when mapped onto Student 2’s score pattern.

- It indicates that Student 1’s scores follow a similar distribution to Student 2, but at a lower magnitude, reflecting proportional alignment in subject strengths.

# Part B: Matrix Operations 

### Que 1: Form a Students × Subjects Matrix :

```python
# matrix using subject score

matrix = data[['math_score', 'science_score', 'english_score']].values

print("Students × Subjects Matrix \n", matrix)

print("\n Shape of Matrix:", matrix.shape)
```

## Interpretation :

- The Students × Subjects matrix represents all 25,000 students’ scores across three subjects (math, science, English), forming a large dataset for analysis.

- Its shape (25000, 3) confirms there are 25,000 rows (students) and 3 columns (subjects), making it ideal for applying linear algebra operations, comparisons, and statistical insights.

## Que 2 : Matrix addition and multiplication :


```python
matrix_addition = matrix + matrix

print("Matrix Addition :\n", matrix_addition)

print()

matrix_multiplication = matrix.T @ matrix

print("Matrix Multiplication: \n", matrix_multiplication)
```

## Interpretation :

- The matrix addition simply doubles each student’s subject scores, showing scaled values rather than new insights.

- The matrix multiplication (matrix.T @ matrix) produces a 3×3 matrix capturing correlations and combined strengths between math, science, and English scores across all 25,000 students — essentially a subject-level similarity/relationship matrix.

## Que 3 : Transpose and Inverse (if possible) :

```python
transpose_matrix = matrix.T

print("Transpose_Matrix: \n", transpose_matrix)

print("\n Shape of Matrix:", matrix.shape)
```

## Interpretation :
- The transpose matrix flips rows and columns, so instead of students × subjects, you now see subjects × students — each row represents all students’ scores in one subject.

- Its shape remains (25000, 3) for the original matrix, but the transpose becomes (3, 25000), making it useful for subject-wise analysis across the entire student dataset.

#### ⚠️ Important: Your matrix is 100 × 3, which is not a square matrix, so it cannot be inverted directly. First create a 3 × 3 covariance matrix.

```python
# Inverse:

cov_matrix = np.cov(matrix.T)
print("Covariance Matrix:\n ",cov_matrix )

# inverse matrix:

inverse_matrix = np.linalg.inv(cov_matrix)

print("\ninverse_matrix:\n", inverse_matrix)
```

## Interpretation :

- The covariance matrix captures how math, science, and English scores vary together across all students — the large positive values show strong correlations among the subjects.

- The inverse matrix provides a way to analyze conditional relationships; the small off-diagonal negatives indicate slight trade-offs, while the positive diagonals highlight each subject’s independent contribution to overall variance.

## Que 4: Determinant :

```python
determinant = np.linalg.det(cov_matrix)

print("Determinant =", determinant)
```

## Interpretation :

- The determinant value (≈9.86 million) shows the covariance matrix is invertible, meaning subjects contribute unique variance.

- It confirms independence of subjects — math, science, and English scores are correlated but not linearly dependent.

# Part C: Linear Transformations & Geometry :

### Que 1: Explain line, plane, and hyperplane with respect to your dataset dimensions :


## Explanation

**Line (1D):**

- A line is a one-dimensional geometric object. If the dataset contains only one feature, such as math_score, then each student's score can be represented on a single line.

**Plane (2D):**

- A plane is a two-dimensional space formed using two features, such as math_score and science_score. In this project, each student is represented as a point on a 2D scatter plot.

**3D Space:**

-When three features (math_score, science_score, and english_score) are used, each student is represented as a point in a three-dimensional coordinate system.

**Hyperplane:**

- A hyperplane exists in four or more dimensions. Since our dataset contains only three numerical features, a true hyperplane cannot be visualized. If additional features such as attendance, computer score, or study hours were added, the data would lie in a higher-dimensional space called a hyperplane.

# Part D: Eigenvalues & Decomposition :

## Que 1 : Compute the eigenvalues and eigenvectors of the covariance matrix :

```python

# Eigenvalues and Eigenvectors of Covariance Matrix
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:\n", eigen_values)


print("\nEigenvectors:\n", eigen_vectors)
```

## Interpretation :
- The eigenvalues show variance strength: one large (≈1120) dominates, meaning most score variation lies along a single main direction.

- The eigenvectors define those directions: the first combines all subjects (overall performance), while the smaller ones capture contrasts between them.

## Que 2: Perform LU Decomposition of the dataset matrix :

```python


# LU Decomposition
P, L, U = lu(cov_matrix)

print("Permutation Matrix (P):\n", P)

print("\nLower Triangular Matrix (L)\n:", L)


print("\nUpper Triangular Matrix (U):\n",U)
```

## Interpretation :

- The LU decomposition breaks the covariance matrix into lower (L) and upper (U) triangular matrices, with a permutation matrix (P) showing no row swaps here.

- The lower triangular matrix captures multipliers used during elimination, while the upper triangular matrix holds the transformed coefficients.

- This factorization is useful for efficient solving of linear systems and numerical stability in matrix computations.

### Que 3 :Perform Singular Value Decomposition (SVD) and explain its role in dimensionality reduction :

```python

# Singular Value Decomposition
U, S, VT = np.linalg.svd(matrix)

print("U Matrix:\n", U)


print("\nSingular Values:\n", S)


print("\nV Transpose Matrix:\n", VT)
```

## Interpretation :
- The singular values show variance strength: one large value dominates, meaning most score information lies along a single main axis.

- The U matrix represents student-specific directions, while the V transpose defines subject-based directions.

- Overall, SVD is great for dimensionality reduction and uncovering hidden score patterns.

# Part E: Dimensionality Reduction :

### Que 1 :Apply Principal Component Analysis (PCA) to reduce the dataset from multiple subjects to 2 dimensions :

```python
from sklearn.decomposition import PCA

X = data[['math_score', 'science_score', 'english_score']]

# Apply PCA
pca = PCA(n_components=2)

X_pca = pca.fit_transform(X)

print("First 5 PCA Components:\n", X_pca[:5])
```

## Interpretation :
- The PCA transformation reduces the 3 subject dimensions (math, science, English) into 2 principal components.

- The first component captures the largest variance — essentially overall performance across subjects.

- The second component captures smaller contrasts, highlighting differences between subject strengths.

```python
# Visualize PCA



plt.figure(figsize=(8,6))

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    color='blue',
    alpha=0.6
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA - Student Performance")

plt.grid(True)

plt.show()
```

### Que 2: Apply Linear Discriminant Analysis (LDA) to classify students into “Above Average” and “Below Average” categories:

```python
# Step 1: Create Categories


# Create target labels
data['Category'] = np.where(
    data['overall_score'] >= data['overall_score'].mean(),
    'Above Average',
    'Below Average'
)

print(data[['overall_score', 'Category']].head())

# Step 2: Encode Labels

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y = encoder.fit_transform(data['Category'])

# Step 3: Select Features

X = data[['math_score',
          'science_score',
          'english_score']]

# Step 4: Apply LDA

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=1)

X_lda = lda.fit_transform(X, y)

print("\nLDA Fit Transform:\n", X_lda[:10])

# Step 5 Visualize LDA

plt.figure(figsize=(8,5))

plt.scatter(
    X_lda,
    np.zeros(len(X_lda)),
    c=y,
    cmap='coolwarm',
    alpha=0.7
)

plt.xlabel("Linear Discriminant")
plt.title("LDA Classification of Students")

plt.yticks([])

plt.show()
```

## Interpretation :
- The category creation step splits students into “Above Average” and “Below Average” groups based on their overall score relative to the mean.

- The LDA transformation projects the 3 subject scores into a single discriminant axis that best separates these two groups.

- The visualization shows clear clustering along the linear discriminant, with red and blue points distinguishing performance categories — making classification patterns easy to interpret.
